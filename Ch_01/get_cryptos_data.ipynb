```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
import pandas as pd
```

```python
# /simple/price endpoint with the required parameters
#cg.get_price(ids='bitcoin', vs_currencies='usd')

```

```python
coins=pd.DataFrame(cg.get_coins_list())
#coins=coins.head()
coins
```

```python
id=coins['id'].values.tolist()
```

```python
# Python program to convert a list 
# to string using join() function 
	
# Function to convert 
def listToString(s): 
	
	# initialize an empty string 
	str1 = " "
	
	# return string 
	return (str1.join(s)) 


import time
from datetime import timezone
import datetime
start_date=datetime.datetime(2014,1,1,tzinfo=timezone.utc) 
end_date=datetime.datetime(2020,8,31, tzinfo=timezone.utc)

df=pd.DataFrame(columns=['datetime','id','name','symbol','price','volume','market_cap'])




for i in (id):

    df1=pd.DataFrame(columns=['datetime','id','name','symbol','price','volume','market_cap'])
    try:
               
        price=[i[1] for i in cg.get_coin_market_chart_range_by_id(i,'usd',
                        time.mktime(start_date.timetuple()),
                        time.mktime(end_date.timetuple()))['prices']]
        market_cap=[i[1] for i in cg.get_coin_market_chart_range_by_id(i,'usd',
                        time.mktime(start_date.timetuple()),
                        time.mktime(end_date.timetuple()))['market_caps']]
        
        total_volume=[i[1] for i in cg.get_coin_market_chart_range_by_id(i,'usd',
                        time.mktime(start_date.timetuple()),
                        time.mktime(end_date.timetuple()))['total_volumes']]
        
        date=[i[0] for i in cg.get_coin_market_chart_range_by_id(i,'usd',
                        time.mktime(start_date.timetuple()),
                        time.mktime(end_date.timetuple()))['prices']]
        name=listToString(coins.loc[coins["id"] == i, 'name'].values)
            
        symbol=listToString(coins.loc[coins["id"] == i, 'symbol'].values)
        
        df1['datetime']= date
        df1['price']=price
        df1['id']=i
        df1['market_cap']=market_cap
        df1['volume']=total_volume
        df1['name']=name
        df1['symbol']=symbol
                                                           
        
    except:
      
        continue
    df=df.append(df1)
```

```python
df['datetime']= pd.to_datetime(df.datetime,  unit = 'ms')
```

```python
compression_opts = dict(method='zip',

                        archive_name='all_cryptos.csv')  

df.to_csv('all_cryptos.zip', index=False,

          compression=compression_opts)  
```
